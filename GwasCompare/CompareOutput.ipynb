### Compare

This loads in the differences between the residualised GWAS and non-residualised GWAS and then compares the two
coefficients


```python
from miscSupports import open_setter, decode_line, terminal_time
from csvObject import write_csv
from pathlib import Path

no_res = r"Y:\data\ukbiobank\software\gwas_pipeline\dev\release_candidate\data\phenotypes\ow18390\output\SKINCOLOUR\8f576a1f-4fd7-4871-854d-a3cbdcb9c816\SKINCOLOUR_imputed.txt.gz"
# no_res = r"Y:\data\ukbiobank\software\gwas_pipeline\dev\release_candidate\data\phenotypes\ow18390\output\HAIRCOLOUR\4c5cf4e7-459d-429a-aeca-1b7a472c3681\HAIRCOLOUR_imputed.txt.gz"
no_res = Path(no_res)

res = r"Y:\data\ukbiobank\software\gwas_pipeline\dev\release_candidate\data\phenotypes\ow18390\output\SKINCOLOUR_RES\4bcfe3ea-921d-4dd1-90ff-e9bc96ec4aad\SKINCOLOUR_RES_imputed.txt.gz"
# res = r"Y:\data\ukbiobank\software\gwas_pipeline\dev\release_candidate\data\phenotypes\ow18390\output\HAIRCOLOUR_RES\e732c936-fae5-4362-85f8-977703cce21f\HAIRCOLOUR_RES_imputed.txt.gz"
res = Path(res)

headers = ["SNP", "CHR", "BP", "BETA_NO_RES", "SE_NO_RES", "P_NO_RES", "BETA_RES", "SE_RES", "P_RES"]
```

```python
no_res_dict = {}

with open_setter(no_res)(no_res) as file:

    print(file.readline())
    for index, line_byte in enumerate(file):
        if index % 100000 == 0:
            print(F"Processed {index}")

        row = decode_line(line_byte, True)
        no_res_dict[row[0]] = {"CHR": row[1], "BP": row[2], "BETA": row[10], "SE": row[11], "P": row[13]}

print(terminal_time())
```

```python
res_dict = {}

with open_setter(res)(res) as file:

    print(file.readline())
    for index, line_byte in enumerate(file):
        if index % 100000 == 0:
            print(F"Processed {index}")

        row = decode_line(line_byte, True)
        res_dict[row[0]] = {"CHR": row[1], "BP": row[2], "BETA": row[10], "SE": row[11], "P": row[13]}

print(terminal_time())
```

```python
out_rows = []
for key, val_dict in no_res_dict.items():
    try:
        out_rows.append([key] + list(val_dict.values()) + list(res_dict[key].values())[2:])
    except KeyError:
        pass

print(terminal_time())

write_csv("I:\Work\Genetics\Residuals\Compare two", "SkinCompare", headers, out_rows)
print(terminal_time())
```
